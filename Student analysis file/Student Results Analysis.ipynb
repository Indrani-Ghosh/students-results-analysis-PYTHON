```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("studentdataset.csv")
df.head(10)
```

# Dropped unnamed values

```python
df = df.drop("Unnamed: 0", axis = 1)
df.head()
```

# Change weekly study hours columns.

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("10-May","5-10")
df.head()
```

# Gender Distribution

```python
ax = sns.countplot(x = "Gender", data = df)
ax.bar_label(ax.containers[0])
sns.set(rc={'figure.figsize':(4,5)})
plt.title("Gender Distribution")
plt.show()
```

FROM THE ABOVE CHART WE HAVE ANALYSED THAT FEMALES ARE MORE THAN THE MALES.

```python
ax = sns.countplot(x = "TestPrep", data = df)
ax.bar_label(ax.containers[0])
sns.set(rc={'figure.figsize':(4,6)})
plt.title("Student Distribution based on testprep course")

plt.show()
```

From the above chart it is analysed that maximum students have not prepared for the test
and less than 10,000 students have completed their test preparation.

# Distribution of Ethnic Groups.

```python
groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
groupB = df.loc[(df["EthnicGroup"] == "group B")].count()
groupC = df.loc[(df["EthnicGroup"] == "group C")].count()
groupD = df.loc[(df["EthnicGroup"] == "group D")].count()
groupE = df.loc[(df["EthnicGroup"] == "group E")].count()
l = ["group A","group B","group C","group D","group E"]
mlist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
plt.pie(mlist, labels = l , autopct = "%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```

```python
ax = sns.countplot(x = "EthnicGroup", data = df)
ax.bar_label(ax.containers[0])
sns.set(rc={'figure.figsize':(7,5)})
plt.title("EthnicGroup Distribution")
plt.show()
```
